```python
df_country=pd.read_excel(r"C:\Users\ayans\OneDrive\Desktop\ML\Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
# plt.pie(country_value, labels=country_name)
```

```python
#Pie Chart:
plt.pie(country_val[:3], labels=country_name[:3], autopct='%1.2f%%')
plt.show()
```

Observation from pie chart: Zomato maximum records or transcation are from India after that USA and United Kingdom

```python
## Numerical Variable
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating counnt'})
```

## Observation
1. When rating is between 4.5 to 4.9 --> Excellent
2. When rating are between 4.0 to 3.4 --> Very good
3. When rating are between 3.5 to 3.9 --> good
4. When rating are between 3.0 to 3.4 --> Avg
5. When rating are between 2.5 to 2.9 --> Avg
6.  When rating are between 2.0 to 2.4 --> Poor

```python
rating.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating counnt", data=rating)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating counnt", hue="Rating color",data=rating)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating counnt", hue="Rating color",data=rating,palette=['black','red','orange','yellow','green','green'])
plt.show()
```

#Observation:
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
##Count Plot:
sns.countplot(x="Rating color", data=rating,palette=['black','red','orange','yellow','green','green'])
plt.show()
```
