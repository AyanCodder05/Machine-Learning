````python
#Q2 (Python – Data Leakage Bug)
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np

X = np.array([[10], [20], [30], [40]])

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test = train_test_split(X_scaled, test_size=0.25)

print(X_test)
````

This is a data leakage issue because the scaler is fit on the entire dataset before splitting, allowing test data information to influence training; the correct approach is to split first, fit the scaler only on the training data, and then transform the test data using the same fitted scaler.

````python
## Fixed Code
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np

X = np.array([[10], [20], [30], [40]])

# Step 1: Split first
X_train, X_test = train_test_split(X, test_size=0.25)

# Step 2: Fit only on train
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Step 3: Transform test
X_test_scaled = scaler.transform(X_test)

print(X_test_scaled)
````

````python
#Q4 (ML – Accuracy Trap)
from sklearn.metrics import accuracy_score

y_true = [1,1,1,1,0]
y_pred = [1,1,1,1,1]

print("Accuracy:", accuracy_score(y_true, y_pred))

# What’s the real problem?

# Model = lazy (always predicts 1)

# Minority class = missed

# Accuracy = fake confidence
````

---

## Wrong Approach (current)

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_true, y_pred))
```

---

##  Correct Fix (use better metrics)

#####  Option 1: Classification Report

```python
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred))
```

---

#####  Option 2: Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_true, y_pred))
```

---

#####  Option 3: Precision, Recall, F1

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
```

---

#####  Why this is the “fix”

* Accuracy hides the problem 
* These metrics **expose the problem** 

Example:

* Recall for class `0` = **0** → model failed completely

---

#####  Interview One-liner

> The fix is not changing accuracy but using metrics like precision, recall, and F1-score to properly evaluate imbalanced data.
