```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])
n = len(x)

# Hyperparameters
alpha = 0.01
iterations = 100

# Initialize parameters randomly
m = np.random.uniform(-10, 10)
c = np.random.uniform(-10, 10)

# Store values for plotting convergence
m_values = []
c_values = []
cost_values = []

# Gradient Descent Loop
for i in range(iterations):
    # Predictions
    y_pred = m * x + c

    # Cost (Mean Squared Error)
    mse = (1/n) * np.sum((y_pred - y)**2)

    # Gradients
    dm = (2/n) * np.sum(x * (y_pred - y))
    dc = (2/n) * np.sum(y_pred - y)

    # Parameter updates
    m -= alpha * dm
    c -= alpha * dc

    # Store for path tracking
    m_values.append(m)
    c_values.append(c)
    cost_values.append(mse)

print("Final Slope (m):", m)
print("Final Intercept (c):", c)
print("Final Cost (MSE):", mse)

# Create grid for visualization
M, C = np.meshgrid(np.linspace(-10, 10, 100), np.linspace(-10, 10, 100))
Z = np.zeros_like(M)

# Compute MSE for each (m, c) pair
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        y_pred_grid = M[i, j] * x + C[i, j]
        Z[i, j] = (1/n) * np.sum((y_pred_grid - y)**2)

# Contour Plot (Cost Surface)
plt.figure(figsize=(8, 6))
contour = plt.contour(M, C, Z, levels=50, cmap='viridis')
plt.clabel(contour, inline=1, fontsize=8)
plt.title("Cost Function Contour (MSE)")
plt.xlabel("Slope (m)")
plt.ylabel("Intercept (c)")

# Plot Gradient Descent Path
plt.plot(m_values, c_values, color='red', marker='o', linewidth=1.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.array([1,2,3,4,5])
x2 = np.array([6,2,7,4,3])
y  = np.array([1,2,5,8,9])
n = len(y)

# build design matrix with bias column 
X = np.column_stack((x1, x2))           
X_b = np.c_[np.ones((n, 1)), X]         

# column vector for y 
y_col = y.reshape(-1, 1)                

#  initialize theta = [c, m1, m2] as column vector 
theta = np.random.uniform(-1, 1, size=(3,1))

# hyperparams
alpha = 0.01
iterations = 1000

# history for diagnostics/plotting 
cost_history = []
theta_history = []

# vectorized gradient descent loop
for it in range(iterations):
    y_pred = X_b @ theta                
    error = y_pred - y_col              
    
    cost = (1/(2*n)) * np.sum(error**2)  
    grad = (1/n) * (X_b.T @ error)       
    
    theta = theta - alpha * grad         
    
    # store for analysis
    cost_history.append(cost)
    theta_history.append(theta.flatten().copy())

# final results
c_final, m1_final, m2_final = theta.flatten()
print("theta (c, m1, m2) =", theta.flatten())
print("Final cost =", cost_history[-1])

#quick plots
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(cost_history)
plt.title("Cost over iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")

plt.subplot(1,2,2)
theta_hist = np.array(theta_history)   # shape (iterations, 3)
plt.plot(theta_hist[:,0], label='c')
plt.plot(theta_hist[:,1], label='m1')
plt.plot(theta_hist[:,2], label='m2')
plt.legend()
plt.title("Parameter paths")
plt.xlabel("Iteration")

plt.tight_layout()
plt.show()
```
