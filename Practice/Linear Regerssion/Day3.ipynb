```python
#Problem: Given a tiny dataset of X = [1, 2, 3, 4] and y = [2, 4, 6, 8], write code from scratch (using only NumPy) to 
#calculate the slope and intercept of the best-fit line using the least squares method. Do not use `sklearn`.
import numpy as np
import pandas as pd

x=np.array([1,2,3,4])
y=np.array([2,4,6,8])

n=len(x)
sum_x=np.sum(x)
sum_y=np.sum(y)
sum_product=np.sum(x*y)
print("Total ofx:",sum_x)
print("Total ofy:",sum_y)
print("Total of product:",sum_product)

slope=((n*sum_product)-(sum_x*sum_y))/((n*np.sum(x**2))-(sum_x)**2)
intercept=((sum_y)-(slope*sum_x))/n
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
#2. Problem: Use `sklearn` to fit a linear regression model to the dataset from Problem 
#1. Print the coefficient and intercept and confirm they match your manual calculation. 
from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([[1], [2], [3], [4]])   
y = np.array([2, 4, 6, 8])

reg = LinearRegression().fit(x, y)
print("Coefficient:", reg.coef_)
print("Intercept:", reg.intercept_)
```

```python
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
# y = 1 * x_0 + 2 * x_1 + 3
y = np.dot(X, np.array([1, 2])) + 3
reg = LinearRegression().fit(X, y)
reg.score(X, y)
reg.coef_
reg.intercept_
reg.predict(np.array([[3, 5]]))
```

```python
#3. Simple Real-World Predictor Problem: You have data on hours studied and exam scores: `hours = [1, 2, 3, 4, 5]`, 
#scores = [50, 55, 70, 75, 85]`. Train a model to predict the score for someone who studied for 2.5 hours. Plot the data points and the regression line.
hours=np.array([[1],[2],[3],[4],[5]])
scores=np.array([50,55,70,75,85])
reg=LinearRegression().fit(hours,scores)

slope=reg.coef_
intercept=reg.intercept_

predicate=(slope*2.5)+intercept
print("Slope: ",slope)
print("Intercept: ",intercept)
print(predicate)
```
