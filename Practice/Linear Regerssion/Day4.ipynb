```python
#4. The "Constant" Predictor Problem: Fit a model to the dataset: `X = [10, 20, 30, 40]`, `y = [5, 5, 5, 5]`. What are
# the slope and intercept? What does this tell you about the relationship between X and y?
# Goal: Understand that a model will still "work" even when there is no real
# relationship.

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

x=np.array([[10],[20],[30],[40]])
y=np.array([5,5,5,5])

model=LinearRegression()
model.fit(x,y)

print("Slope: ",model.coef_[0])
print("Intercept: ",model.intercept_)
```

```python
# 5. Single Feature from a DataFrame Problem: Load the `diabetes` dataset from `sklearn.datasets`. Create a pandas
# DataFrame. Use only the 'bmi' feature to predict the 'target' (disease progression). What is
# the R-squared value of your model?
# Goal: Get comfortable working with DataFrames and a standard library dataset.

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score



diabetes = load_diabetes()


col=diabetes.feature_names
df=pd.DataFrame(diabetes.data, columns=col)
print("\n+++++++\n")
print("Data Frame:\n",df)
print("\n+++++++\n")
df["target"]=diabetes.target

x=df[["bmi"]]
y=df["target"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

print("Train x\n", x_train)
print("\n+++++++\n")
print("Test x\n", x_test)
print("\n+++++++\n")
print("Train y\n", y_train)
print("\n+++++++\n")
print("Test y\n", y_test)
print("\n+++++++\n")

model.fit(x_train, y_train)

y_pred=model.predict(x_test)
r_square=r2_score(y_test, y_pred)

print("Predicated y:",y_pred)
print("R-squared value: ",r_square)
```
